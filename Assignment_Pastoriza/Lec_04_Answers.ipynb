$\textbf{1. Kruskal's Algorithm}$

Kruskal's algorithm finds a minimum spanning forest of an undirected edge-weighted graph. If the graph is connected, it finds a minimum spanning tree. It is a greedy algorithm that in each step adds to the forest the lowest-weight edge that will not form a cycle.

Step 1: Find all minimum edge without forming a cycle.

$I - A$ (weight of 5)

$A - C$ (weight of 10)

$C - E$ (weight of 15)

$C - B$ (weight of 20)

$B - E$ (weight of 25): This forms a cycle with $C - B$ and $C - E$, so we skip this.

$C - D$ (weight of 30)

$E - F$ (weight of 35)

$I - E$ (weight of 40): This forms a cycle with $I - A$, $A - C$ and $C - E$, so we skip this.

$E - G$ (weight of 45)

Total cost: 160

Since we only have 8 vertices we stop here because of absolute rule spanning trees such that, $|e| = |v| - 1$.

$\textbf{Hence, this spanning tree has the lowest-cost edge, proven by Kruskal's Algorithm.}$

```python

import networkx as nx
import matplotlib.pyplot as plt

# Create the graph
G = nx.Graph()
G.add_nodes_from(["A","B","C","D","E","F","G","I"])
edgelist = [("A","C"),("A","I"),("C","B"),("C","E"),("C","D"),("B","E"),("I","E"),("E","F"),("F","G"),("G","E")]
G.add_edges_from(edgelist)

# Define the path you provided
path = [("I", "A"),("A", "C"), ("C", "E"),("B", "C"),("D", "C"), ("F", "E"),("G", "E")]

# Generate the minimum spanning tree with the edges of the path highlighted
MST = nx.Graph()
for edge in path:
    MST.add_edge(*edge)

# Plot the graph with highlighted edges
plt.figure(figsize=(10, 8))
pos = nx.planar_layout(G)  # Seed for reproducible layout
nx.draw_networkx_edge_labels(
    G, pos,
    edge_labels={("A","C"):10,("A","I"):5,("C","B"):20,("C","E"):15,("C","D"):30,("B","E"):25,("I","E"):40,("E","F"):35,("F","G"):50,("G","E"):45},
    font_color='red'
)
nx.draw(G, pos=pos, with_labels=True)
nx.draw_networkx_edges(MST, pos, edge_color='violet', width=2)  # Add MST edges in blue
plt.title("Kruskal's Algorithm")
plt.show()
```

$\textbf{2. Prim's Algorithm}$

Prim's algorithm is a greedy algorithm that finds a minimum spanning tree for a weighted undirected graph. This means it finds a subset of the edges that forms a tree that includes every vertex, where the total weight of all the edges in the tree is minimized. It is very similar to Kruskal's method, the only difference is that when selecting minimum edges, it should be connected to the current tree.

Step 1: From the graph, we select the minimum cost edge.

That would be $I \rightarrow A$, which has a weight of 5.

Step 2: From now on, we select all the other edges with the minimum cost from the graph. However, it should be connected to the previously selected vertices and should not form a cycle, if not it would not be a spanning tree.


$I - A$ (weight of 5)

$A - C$ (weight of 10)

$C - E$ (weight of 15)

$C - B$ (weight of 20)

$B - E$ (weight of 25): This forms a cycle with $C - B$ and $C - E$, so we skip this.

$C - D$ (weight of 30)

$E - F$ (weight of 35)

$I - E$ (weight of 40): This forms a cycle with $I - A$, $A - C$ and $C - E$, so we skip this.

$E - G$ (weight of 45)

Total cost: 160

$\textbf{Hence, this spanning tree has the lowest-cost edge, proven by Prim's Algorithm.}$

```python

import networkx as nx
import matplotlib.pyplot as plt

# Create the graph
G = nx.Graph()
G.add_nodes_from(["A","B","C","D","E","F","G","I"])
edgelist = [("A","C"),("A","I"),("C","B"),("C","E"),("C","D"),("B","E"),("I","E"),("E","F"),("F","G"),("G","E")]
G.add_edges_from(edgelist)

# Define the path you provided
path = [("I", "A"),("A", "C"), ("C", "E"),("B", "C"),("D", "C"), ("F", "E"),("G", "E")]

# Generate the minimum spanning tree with the edges of the path highlighted
MST = nx.Graph()
for edge in path:
    MST.add_edge(*edge)

# Plot the graph with highlighted edges
plt.figure(figsize=(10, 8))
pos = nx.planar_layout(G)  # Seed for reproducible layout
nx.draw_networkx_edge_labels(
    G, pos,
    edge_labels={("A","C"):10,("A","I"):5,("C","B"):20,("C","E"):15,("C","D"):30,("B","E"):25,("I","E"):40,("E","F"):35,("F","G"):50,("G","E"):45},
    font_color='red'
)
nx.draw(G, pos=pos, with_labels=True)
nx.draw_networkx_edges(MST, pos, edge_color='green', width=2)  # Add MST edges in blue
plt.title("Prim's Algorithm")
plt.show()
```

$\textbf{3. Djikstra's Shortest Path Algorithm}$

Dijkstra's Algorithm finds the shortest path between a given node (which is called the "source node") and all other nodes in a graph. This algorithm uses the weights of the edges to find the path that minimizes the total distance (weight) between the source node and all other nodes.

Suppose that our source is $I(G)$ and destination is $F(G)$.

Step 1: Find all possible routes.

Case 1: $I \rightarrow E \rightarrow F$

Total distance: 75

Case 2: $I \rightarrow E \rightarrow G \rightarrow F$

Total distance: 135

Case 3: $I \rightarrow A \rightarrow C \rightarrow E \rightarrow F$

Total distance: 65

Case 4: $I \rightarrow A \rightarrow C \rightarrow E \rightarrow G \rightarrow F$

Total distance: 125

Case 5: $I \rightarrow A \rightarrow C \rightarrow B \rightarrow E \rightarrow  F$

Total distance: 95

Case 6: $I \rightarrow A \rightarrow C \rightarrow B \rightarrow E \rightarrow G \rightarrow F$

Total distance: 155

Step 2: Select shortest path from all possible routes.



```python

import networkx as nx
import matplotlib.pyplot as plt

# Create the graph
G = nx.Graph()
G.add_nodes_from(["A","B","C","D","E","F","G","I"])
edgelist = [("A","C"),("A","I"),("C","B"),("C","E"),("C","D"),("B","E"),("I","E"),("E","F"),("F","G"),("G","E")]
G.add_edges_from(edgelist)

# Define the path you provided
path = [("I", "A"),("A", "C"), ("C", "E"),("E", "F")]

# Generate the minimum spanning tree with the edges of the path highlighted
MST = nx.Graph()
for edge in path:
    MST.add_edge(*edge)

# Plot the graph with highlighted edges
plt.figure(figsize=(10, 8))
pos = nx.planar_layout(G)  # Seed for reproducible layout
nx.draw_networkx_edge_labels(
    G, pos,
    edge_labels={("A","C"):10,("A","I"):5,("C","B"):20,("C","E"):15,("C","D"):30,("B","E"):25,("I","E"):40,("E","F"):35,("F","G"):50,("G","E"):45},
    font_color='red'
)
nx.draw(G, pos=pos, with_labels=True)
nx.draw_networkx_edges(MST, pos, edge_color='red', width=2)  # Add MST edges in blue
plt.title("Djikstra's Shortest Path Algorithm")
plt.show()
```
