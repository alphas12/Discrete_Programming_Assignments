2. Prove Minkowski's Inequality for sums, $\forall \ p>1, (a_k, b_k)>0$:
\begin{gather}
\begin{bmatrix}
\Sigma_{k=1}^n |a_k + b_k|^p
\end{bmatrix}^\frac{1}{p} \leq \begin{bmatrix} \Sigma_{k=1}^n |a_k|^p \end{bmatrix}^\frac{1}{p} + \begin{bmatrix} \Sigma_{k=1}^n |b_k|^p \end{bmatrix}^\frac{1}{p}
\end{gather}


$\textbf{Proving:}$

Step 1: Definition of $l^P$ Space:
Given sequences $a = (a_1, a_2, a_3, \ldots)$ and $b = (b_1, b_2, b_3, \ldots)$ belonging to $l^P$ space, which implies $a$, and $b$ are sequences whose $p$-th power of the absolute values of their elements are summable. This implies:


- $(a, b) \in l^P$
- $b = (b_1, b_2, b_3, \ldots)$
- $a + b = (a_1 + b_1, a_2 + b_2, \ldots)$
- $a+b \in l^p a\in l^P b\in l^P$


Step 2: Using the Triangle Inequality

By the triangle inequality:
- $|a_k + b_k| \leq |a_k |+ |b_k| \implies |a_k + b_k|^p \leq (|a_k| + |b_k|)|a_k + b_k|^{p-1}, where \; \forall k \in \mathbb{N}$

From this inequality, we obtain:

- $\Sigma_{k=1}^n |a_k + b_k| ^ p \leq\Sigma_{k=1}^n (|a_k| + |b_k|)|a_k + b_k|^{p-1}$


Step 3: Applying the Holder's Inequality

- $\Sigma_{k=1}^n |a_k + b_k| ^ p \leq \Sigma_{k=1}^n |a_k||a_k + b_k|^{p-1} + \Sigma_{k=1}^n |b_k||a_k + b_k|^{p-1}$

where $a = (a_1, a_2, a_3, \ldots) \in l^u$  and $b = (b_1, b_2, b_3, \ldots) \in l^v$

Step 4: Conjugate
$\frac{1}{u} + \frac{1}{v} = 1$ where u > 1. By Holder's inequality, we have the following:

- $\implies \Sigma_{i=1}^\infty|a_i + b_i| \leq (\Sigma_{j=1}^\infty |a_j|^u)^{\frac{1}{u}} + (\Sigma_{k=1}^\infty |b_k|v)^{\frac{1}{v}} $

Let's define:
- $\overline a = ({a_1, a_2, \ldots, a_n, 0,0,0})$

- $\overline b = ({b_1, b_2, \ldots, a_n, 0,0,0})$

Step 5: Expanded Form:

- $ \Sigma_{i=1}^\infty|a_i + b_i| = a_1 \cdot b_1 + a_2 \cdot b_2 + \ldots  + a_n \cdot b_n$

- $(\Sigma_{j=1}^\infty |a_j|^u)^{\frac{1}{u}}  = (|a_1|^u + |a_2|^u + \ldots  + |a_n|^u  + |a_n + 1|^u )^{\frac{1}{u}} = (\Sigma_{j=1}^\infty |a_j|^u)^{\frac{1}{u}}$

- $(\Sigma_{k=1}^\infty |b_k|^u)^{\frac{1}{v}}  = (|b_1|^v + |b_2|^v + \ldots  + |b_n|^v + |b_n + 1|^v )^{\frac{1}{v}} = (\Sigma_{k=1}^\infty |b_k|^u)^{\frac{1}{v}}$



Step 6: Minskowski's Inequality for Sums:

Given $p > 1$, we aim to prove:
\begin{gather}
\begin{bmatrix}
\Sigma_{k=1}^n |a_k + b_k|^p
\end{bmatrix}^\frac{1}{p} \leq \begin{bmatrix} \Sigma_{k=1}^n |a_k|^p \end{bmatrix}^\frac{1}{p} + \begin{bmatrix} \Sigma_{k=1}^n |b_k|^p \end{bmatrix}^\frac{1}{p}
\end{gather}

For the case $p = 1$, we have:

\begin{gather}
\begin{bmatrix}
\Sigma_{k=1}^n |a_k + b_k|^1
\end{bmatrix}^\frac{1}{1} \leq \begin{bmatrix} \Sigma_{k=1}^n |a_k|^1 \end{bmatrix}^\frac{1}{1} + \begin{bmatrix} \Sigma_{k=1}^n |b_k|^1 \end{bmatrix}^\frac{1}{1}


\newline \Sigma_{k=1}^\infty |a_k + b_k| = \Sigma_{k=1} ^\infty|a_k| + |b_k|
\end{gather}

Since $|a_k + b_k| \leq |a_k |+ |b_k| \forall k \in \mathbb{N}$, we have: 

- $ \Sigma_{k=1}^n |a_k + b_k| = \Sigma_{k=1} ^n|a_k| + |b_k|$

Taking the limit as $ n$ to $ \infty$, we get:

- $\lim _ {n \to \infty} \Sigma_{k=1}^n |a_k + b_k| \leq lim _ {n \to \infty} \Sigma_{k=1}^n |a_k| + |b_k|$

For $p > 1$, Using the Holders inequality:

- $\Sigma_{k=1}^n |a_k + b_k| ^ p \leq \Sigma_{k=1}^n |a_k||a_k + b_k|^{p-1} + \Sigma_{k=1}^n |b_k||a_k + b_k|^{p-1}$

This leads to:
\begin{gather}
\begin{bmatrix}
\Sigma_{k=1}^n |a_k + b_k|
\end{bmatrix} \leq \begin{bmatrix} \Sigma_{k=1}^n |a_k|^p \end{bmatrix}^\frac{1}{p} + \begin{bmatrix} \Sigma_{k=1}^n |b_k|^q \end{bmatrix}^\frac{1}{q}
\end{gather}

Conjugate 

$\frac{1}{p} + \frac{1}{q} = 1$

Considered that $a_i = a_k$ and $b_i = (a_k + b_k)^{p-1}$ where:

\begin{gather}

Step 7: Transitivite Property

In this proof, it utilize the transitive property to combine two inequalities derived from Hölder's inequality into one single inequality. 

Initially, we have two inequalities obtained from Hölder's inequality:

- $\begin{bmatrix}
\Sigma_{i=1}^n |a_i + (a_i + b_i)^{p-1}|
\end{bmatrix} \leq \begin{bmatrix} \Sigma_{j=1}^n |a_j|^p \end{bmatrix}^\frac{1}{p} + \begin{bmatrix} \Sigma_{k=1}^n |(a_k + b_k)^{p}|\end{bmatrix}^\frac{1}{q}$


- $\begin{bmatrix}
\Sigma_{i=1}^n |b_i + (a_i + b_i)^{p-1}|
\end{bmatrix} \leq \begin{bmatrix} \Sigma_{j=1}^n |b_j|^p \end{bmatrix}^\frac{1}{p} + \begin{bmatrix} \Sigma_{k=1}^n |(a_k + b_k)^{p}|\end{bmatrix}^\frac{1}{q}$


To combine these inequalities, we observe that both of them have the term $\begin{bmatrix}\Sigma_{k=1}^n |(a_k + b_k)^{p}\end{bmatrix}^\frac{1}{q}$. By the transitive property of inequalities, we can conclude that the sum of the terms inside the brackets on the left-hand sides of the two inequalities is less than or equal to the sum of the terms inside the brackets on the right-hand sides:

- $\begin{bmatrix}\Sigma_{k=1}^n |a_k + b_k|^{p}\end{bmatrix}^\frac{1}{q} + \begin{bmatrix}\Sigma_{k=1}^n |a_k + b_k|^{p}\end{bmatrix}^{\frac{1}{q}}$

is less than or equal to:

- $\begin{bmatrix}\Sigma_{j=1}^n |a_j|^p\end{bmatrix}^{\frac{1}{p}} + \begin{bmatrix}\Sigma_{j=1}^n |b_j|^p \end{bmatrix}^{\frac{1}{p}}$

This process allows us to combine the two inequalities into one, which is a crucial step in deriving Minkowski's inequality for sums.

Step 8: Combined Inequality

By applying the transitive property, we obtain combined inequality:


\begin{gather}
\frac{\Sigma_{i=1}^n |a_i + b_i| ^ p}{(\Sigma_{k=1}^n |a_k + b_k|^p)^\frac{1}{q}}

$\frac{a^k}{a^l} = a^{k-l}$

$\textbf{Proof of the Triangle Inequality:}$

Let's considered the following cases:

Case 1: Suppose $(x,y) \geq 0$. Then, since $x + y \geq 0$ we have:

- $|x| = x$
- $|y| = y$
- $|x| + |y| = x + y$
- $|x + y| = x + y$

Thus, $|x| + |y| = |x + y|$ are both equal, and the inequality is satisfied.


Case 2: $(x,y) < 0$

If both $x$ and $y$ are negative, the inequality holds trivially as both sides are equal:

- $|x| = -x$
- $|y| = -y$
- $|x| + |y| = -x + -y$
- $|x + y| = -(x + y)$

In this case, both sides are negative, and we need to show that
$-x - y \leq -(x + y).$

Starting with $-x - y$:

$-x - y = -(x + y)$

Hence, $|x| + |y| = |x + y|$ are both equal, and the inequality is satisfied.

Case 3: $ x \geq 0, y < 0 $, then we have:

- $|x| = x$
- $|y| = -y$
- $|x| + |y| = x + (-y)$
- $|x + y| = x - y$

In this case, $|x| + |y| = |x + y|$ are both equal.

Case 4: $x < 0, y \geq 0$ , then we have:

- $|x| = -x$
- $|y| = y$
- $|x| + |y| = y - x$
- $|x + y| = -(x-y)$

This case is true, since $|x| + |y| = |x + y|$ are both equal.

$\textbf{Conclusion}$

In conclusion, the proof establishes the triangle inequality $|x + y| \leq |x| + |y|, \forall (x, y) \in \mathbb{R}$ by considering different cases.

$$

\frac{(\sum_{i=1}^{n}2a_i)^{2}}{\sum_{i=1}^{n}2b_i} \leq \sum_{i=1}^{n} \frac{(2a_i)^{2}}{2b_i}

$$

$$
\frac{2(\sum_{i=1}^{n}a_i)^{2}}{\sum_{i=1}^{n}b_i} \leq 2\space[\sum_{i=1}^{n} \frac{(a_i)^{2}}{b_i}]

$$

$\textbf{Conclusion}$

This is the desired inequality, so we have proven that: 


$$

\frac{(\sum_{i=1}^{n}u_i)^{2}}{\sum_{i = 1}^{n}v_i} \leq \sum_{i=i}^{n} \frac{(u_i)^{2}}{v_i}
