$\textbf{Graph}$
-

A structure used to model pairwise relationships objects. It consists of $\textbf{vertices}$, known as the nodes, and the $\textbf{edges}$, links between the vertices. There are two types of graph:

$\textbf{Undirected Graph}$
-

In an undirected graph, edges have no direction. This
means that if there's an edge between vertex A and vertex B, it implies a connection 
from B to A as well. For example in this graph below, if vertex 6 is connected to 
vertex 0, it also implies that vertex 0 is connected to vertex 6.

```python
import matplotlib.pyplot as plt
import networkx as nx

n = 9 
m = 6 
seed = 60

G = nx.gnm_random_graph(n, m, seed=seed)

edgelist = [(2, 2), (5, 5)]
G.add_edges_from(edgelist)

pos = nx.spring_layout(G, seed=seed)

plt.figure(figsize=(8, 6))

edge_colors = []
for edge in G.edges():
    if edge[0] == edge[1]:
        edge_colors.append('red')
    else:
        edge_colors.append('blue') 

nx.draw(G, pos=pos, with_labels=True, edge_color=edge_colors, node_color='skyblue', node_size=500)

plt.show()
```

$\textbf{Bipartite Graph}$
-

A bipartite graph is a type of graph where the set of vertices can be divided into two disjoint sets $U$ and $V$ such that every edge connects a vertex in $U$ to a vertex in 
$V$. There are no edges connecting vertices within the same set. This type of graph is often used to model relationships between two different classes of objects.

```python
import matplotlib.pyplot as plt
import networkx as nx


secret = nx.Graph()
secret.add_nodes_from(['1','2','3'], bipartite=0) # Add the node attribute "bipartite"
secret.add_nodes_from(['4', '5', '6'], bipartite=1)
secret.add_edges_from([('1','4'), ('1','5'), ('1','6'), ('2','4'), ('2', '5'), ('2','6'), ('3', '4'),
                               ('3', '5'), ('3', '6')])

top = nx.bipartite.sets(secret)[0]
pos = nx.bipartite_layout(secret, top)
nx.draw(secret, pos=pos, with_labels=True, node_color=['pink','magenta','red','deepskyblue','cyan', 'yellow'])
plt.show()
```


$\textbf{Subgraph}$
- 

A subgraph is a graph formed from a subset of the vertices and edges of another graph. Specifically, if $G = (V, E)$ is a graph, then a subgraph $H$ of $G$ is a graph such that:

- The vertex set of $H$ is a subset of the vertex set of $G (V(H) \subseteq V(G))$.
- The edge set of $H$ is a subset of the edge set of $G ((E(H) \subseteq E(G)))$.
- Each edge in $E(H)$ connects two vertices in $V(H)$.


```python
import matplotlib.pyplot as plt
import networkx as nx

# Create the original graph
G = nx.Graph()

# Add nodes and edges to the original graph
G.add_nodes_from(range(1, 11))
edges = [
    (1, 2), (1, 3), (2, 4), (2, 5), (3, 6),
    (3, 7), (4, 8), (5, 9), (6, 10), (7, 8),
    (8, 9), (9, 10)
]
G.add_edges_from(edges)

# Define the subgraph
sub_nodes = [1, 2, 4, 5, 8, 9]
H = G.subgraph(sub_nodes)

# Draw the original graph
plt.figure(figsize=(10, 5))
plt.subplot(121)
pos = nx.spring_layout(G, seed=42)
nx.draw(G, pos=pos, with_labels=True, node_color='lightblue', edge_color='gray', node_size=700, font_size=10)
plt.title("Original Graph")

# Draw the subgraph
plt.subplot(122)
pos_sub = {node: pos[node] for node in sub_nodes}
nx.draw(H, pos=pos_sub, with_labels=True, node_color='lightgreen', edge_color='black', node_size=700, font_size=10)
plt.title("Subgraph")

plt.show()
```

$\textbf{Handshake Theorem}$
- 

The Handshake Theorem is a fundamental result in graph theory that states that the sum of the degrees of all vertices in an undirected graph is twice the number of edges in the graph. Mathematically, it can be expressed as:

\begin{gather}
\sum_{v \in V} \deg(v) = 2|E|
\end{gather}

where $deg(v)$ denotes the degree of vertex $v$, $V$ is the set of vertices, and $|E|$ is the number of edges.

$\textbf{Example}$

Consider a simple graph with 5 vertices and 4 edges. Let's verify the Handshake Theorem for this graph.


```python
import matplotlib.pyplot as plt
import networkx as nx

# Create the graph
G = nx.Graph()

# Define the edges
edges = [(1, 2), (1, 3), (1, 4), (2, 3)]

# Add edges to the graph
G.add_edges_from(edges)

# Calculate the degree of each vertex
degrees = dict(G.degree())

# Verify the Handshake Theorem
total_degree = sum(degrees.values())
num_edges = G.number_of_edges()

print(f"Total degree of all vertices: {total_degree}")
print(f"Twice the number of edges: {2 * num_edges}")
print(f"Handshake Theorem holds: {total_degree == 2 * num_edges}")

# Draw the graph
pos = nx.spring_layout(G, seed=42)  # Positions for all nodes

# Draw nodes
nx.draw_networkx_nodes(G, pos, node_size=700, node_color='lightblue')

# Draw edges
nx.draw_networkx_edges(G, pos, width=2)

# Draw labels
nx.draw_networkx_labels(G, pos, font_size=12, font_family='sans-serif')

# Set plot title
plt.title('Graph with Degrees of Vertices')

# Show the plot
plt.axis('off')
plt.show()

```

$\textbf{Walk}$

A walk in a graph is a sequence of vertices and edges where each edge's endpoints are the preceding and following vertices in the sequence. A walk can traverse vertices and edges multiple times. Walks can be of any length, and they may or may not form a loop.

```python
import networkx as nx
import matplotlib.pyplot as plt

# Create a graph
G = nx.Graph()
G.add_nodes_from(['Paris', 'Dusseldorf', 'Berlin', 'Venice', 'Brussels', 'Luxembourg'])

# Define the edges
edgelist = [
    ('Paris', 'Dusseldorf'),
    ('Paris', 'Luxembourg'), 
    ('Dusseldorf', 'Luxembourg'), 
    ('Dusseldorf', 'Berlin'), 
    ('Berlin', 'Venice'), 
    ('Berlin', 'Brussels'),
    ('Berlin', 'Luxembourg'), 
    ('Venice', 'Brussels'),
    ('Brussels', 'Luxembourg')
]
G.add_edges_from(edgelist)

# Define the walk
walk = [
    ('Dusseldorf', 'Berlin'),
    ('Berlin', 'Venice'), 
    ('Venice', 'Brussels'),
    ('Brussels', 'Berlin'), 
    ('Berlin', 'Dusseldorf'), 
    ('Dusseldorf', 'Paris')
]

# Compute positions for visualization
pos = nx.spring_layout(G, seed=42)  # Using spring layout for better visualization

# Draw the graph
nx.draw(G, pos=pos, with_labels=True, node_color='skyblue', node_size=700, font_size=10)
nx.draw_networkx_edges(G, pos=pos, edge_color='gray')  # Draw all edges in gray
nx.draw_networkx_edges(G, pos=pos, edgelist=walk, edge_color='red', width=2)  # Highlight the walk in red

# Display the plot
plt.title('Graph Walk Example')
plt.show()
```

The graph visualization will show nodes representing cities and edges representing connections between them. The specific walk will be highlighted in red, demonstrating the path through the graph according to the defined sequence of edges: $\textbf{Dusseldorf -> Berlin -> Venice -> Brussels -> Berlin -> Dusseldorf -> Paris}$.

$\textbf{Trail}$

A trail in a graph is a sequence of vertices and edges where each edge is distinct and appears exactly once. In other words, a trail does not repeat any edge but can visit vertices multiple times.

```python
import networkx as nx
import matplotlib.pyplot as plt

# Create a graph
G = nx.Graph()
G.add_nodes_from(['Paris', 'Dusseldorf', 'Berlin', 'Venice', 'Brussels', 'Luxembourg'])

# Define the edges
edgelist = [
    ('Paris', 'Dusseldorf'),
    ('Paris', 'Luxembourg'), 
    ('Dusseldorf', 'Luxembourg'), 
    ('Dusseldorf', 'Berlin'), 
    ('Berlin', 'Venice'), 
    ('Berlin', 'Brussels'),
    ('Berlin', 'Luxembourg'), 
    ('Venice', 'Brussels'),
    ('Brussels', 'Luxembourg')
]
G.add_edges_from(edgelist)

# Define the trail
trail = [
    ('Paris', 'Luxembourg'),
    ('Luxembourg', 'Berlin'), 
    ('Berlin', 'Venice'),
    ('Venice', 'Brussels'), 
    ('Brussels', 'Berlin'), 
    ('Berlin', 'Dusseldorf')
]

# Compute positions for visualization
pos = nx.spring_layout(G, seed=42)  # Using spring layout for better visualization

# Draw the graph
nx.draw(G, pos=pos, with_labels=True, node_color='skyblue', node_size=700, font_size=10)
nx.draw_networkx_edges(G, pos=pos, edge_color='gray')  # Draw all edges in gray
nx.draw_networkx_edges(G, pos=pos, edgelist=trail, edge_color='red', width=2)  # Highlight the trail in red

# Display the plot
plt.title('Graph Trail Example')
plt.show()
```

The graph visualization will show nodes representing cities and edges representing connections between them. The specific trail will be highlighted in red, demonstrating the path through the graph according to the defined sequence of edges: $\textbf{Paris -> Luxembourg -> Berlin -> Venice -> Brussels -> Berlin -> Dusseldorf}$.
 This trail illustrates a sequence where each edge is used exactly once, but vertices can be revisited.

A circuit is a closed walk in a graph where no edges are repeated. This means it starts and ends at the same vertex, includes at least one edge, and doesn't traverse any edge more than once

$\textbf{Connectedness a Graphs}$
-

- A graph $G$ is connected if there is a path between any two vertices in the graph. This means that for any pair of vertices $v$ and $w$, there exists a sequence of edges that allows traversal from $v$ to $w$.
- If $G$ is connected, then removing any single edge from a circuit (closed loop) will not disconnect the graph.

```python
import networkx as nx
import matplotlib.pyplot as plt

# Create a graph
G = nx.Graph()

# Add nodes
nodes = ['A', 'B', 'C', 'D', 'E', 'F']
G.add_nodes_from(nodes)

# Add edges to ensure the graph is connected
edges = [('A', 'B'), ('A', 'C'), ('B', 'D'), ('C', 'D'), ('D', 'E'), ('E', 'F'), ('F', 'A')]
G.add_edges_from(edges)

# Visualize the graph
pos = nx.spring_layout(G)  # Positions for all nodes

# Draw the graph
nx.draw(G, pos, with_labels=True, node_color='lightblue', node_size=700, font_size=12)
nx.draw_networkx_edges(G, pos, edgelist=edges, edge_color='gray')

# Show the plot
plt.title('Connected Graph Example')
plt.show()

# Check if the graph is connected
is_connected = nx.is_connected(G)
print(f"The graph is connected: {is_connected}")

# Example of removing an edge from a circuit
G.remove_edge('F', 'A')

# Check if the graph is still connected after removing an edge from the circuit
is_connected_after_removal = nx.is_connected(G)
print(f"The graph is still connected after removing an edge: {is_connected_after_removal}")

# Visualize the graph after edge removal
nx.draw(G, pos, with_labels=True, node_color='lightblue', node_size=700, font_size=12)
nx.draw_networkx_edges(G, pos, edgelist=G.edges, edge_color='red')
plt.title('Graph After Removing an Edge from the Circuit')
plt.show()

```

$\textbf{Properties Demonstrated}$

- The graph is initially connected, demonstrating that there is a path between any pair of vertices.
- Removing an edge from a circuit in a connected graph does not disconnect the graph, illustrating the robustness of connectivity in such graphs.

- 

A connected component of a graph is a subgraph in which any two vertices are connected to each other by paths, and which is connected to no additional vertices in the supergraph.

```python
import matplotlib.pyplot as plt
import networkx as nx

# Define the parameters for the graph
n = 8  # Number of nodes
m = 6  # Number of edges
seed = 55  # Seed for random number generator

# Generate the random graph
G = nx.gnm_random_graph(n, m, seed=seed)

# Add additional edges
edgelist = [(1, 2), (2, 3), (3, 4), (1, 4), (6, 7), (7, 8)]
G.add_edges_from(edgelist)

# Get the connected components
connected_components = list(nx.connected_components(G))

# Plot the graph with connected components highlighted
plt.figure(figsize=(8, 6))
pos = nx.spring_layout(G, seed=seed)
for i, component in enumerate(connected_components):
    nx.draw_networkx_nodes(G, pos, nodelist=component, node_color=f'C{i}', label=f'Component {i+1}')
nx.draw_networkx_edges(G, pos)
plt.title('Graph with Connected Components')
plt.legend()
plt.show()

# Print the connected components
print("Connected Components:")
for i, component in enumerate(connected_components):
    print(f"Component {i+1}: {component}")

```

These connected components highlight the structure of the graph, showing how nodes are grouped together based on their connectivity. Components 1 and 3 are connected internally, but they are not connected to each other, while component 2 is isolated.

$\textbf{Eulerian Circuit}$
- 

An Eulerian circuit is a traversal of a graph that visits every edge exactly once, starting and ending at the same vertex.

- An Eulerian circuit is a path that covers all the edges of a graph, visiting each edge exactly once.
- It starts and ends at the same vertex, forming a closed loop.
- Every vertex in the graph must have an even degree for an Eulerian circuit to exist.
- If a graph is connected and all its vertices have even degrees, it contains at least one Eulerian circuit.


```python
import matplotlib.pyplot as plt
import networkx as nx

# Create a graph representing a house
G = nx.Graph()

# Define the rooms (vertices) and corridors (edges)
rooms = ['Living Room', 'Kitchen', 'Bedroom', 'Bathroom', 'Study']
corridors = [('Living Room', 'Kitchen'), ('Living Room', 'Bedroom'), 
             ('Kitchen', 'Bedroom'), ('Bedroom', 'Bathroom'), ('Bedroom', 'Study')]

# Add rooms and corridors to the graph
G.add_nodes_from(rooms)
G.add_edges_from(corridors)

# Draw the graph
pos = nx.spring_layout(G)
nx.draw(G, pos, with_labels=True, node_color='lightblue', node_size=2000, font_size=12, font_weight='bold')
nx.draw_networkx_edges(G, pos, edge_color='gray', width=2)

# Add title
plt.title("Eulerian Circuit in a House")

# Display the plot
plt.show()

```

In this example, the Eulerian circuit represents a path that starts at the Living Room, visits each room once, and returns to the Living Room. It's like navigating through the house, exploring every room and corridor exactly once. This illustrates the concept of an Eulerian circuit, where every edge (corridor) is traversed once, and the path forms a closed loop.

$\textbf{Eulerian Trail}$
- 
An Eulerian trail in a graph is a path that starts at one vertex and ends at another, traversing every edge exactly once and passing through every vertex at least once. It's like exploring a maze where you need to visit every passage once, but you don't necessarily start and end at the same point.

```python
import matplotlib.pyplot as plt
import networkx as nx

# Create a graph representing a city park
G = nx.Graph()

# Define the landmarks (vertices) and paths (edges)
landmarks = ['Entrance', 'Fountain', 'Playground', 'Gazebo', 'Picnic Area']
paths = [('Entrance', 'Fountain'), ('Fountain', 'Playground'), ('Playground', 'Gazebo'),
         ('Gazebo', 'Picnic Area'), ('Picnic Area', 'Fountain'), ('Gazebo', 'Entrance')]

# Add landmarks and paths to the graph
G.add_nodes_from(landmarks)
G.add_edges_from(paths)

# Draw the graph
pos = nx.spring_layout(G)
nx.draw(G, pos, with_labels=True, node_color='lightgreen', node_size=2000, font_size=12, font_weight='bold')
nx.draw_networkx_edges(G, pos, edge_color='gray', width=2)

# Add title
plt.title("Eulerian Trail in a City Park")

# Display the plot
plt.show()

```

In this example, an Eulerian trail represents a path through the park that starts at the Entrance, visits every landmark once, and ends at another landmark. It's like following a treasure map that leads you through the park, exploring every path and discovering every landmark along the way. This illustrates the concept of an Eulerian trail, where every edge (path) is traversed once, and the path may start and end at different vertices.

$\textbf{Hamiltonian Circuit}$
- 
A Hamiltonian circuit in a graph is a special type of cycle that visits every vertex exactly once, except for the starting and ending vertex, which are the same. It's like taking a tour through a city where you visit every landmark exactly once before returning to your starting point.

```python
import matplotlib.pyplot as plt
import networkx as nx

# Create a graph representing a city with landmarks
G = nx.Graph()

# Define the landmarks (vertices) and connections (edges)
landmarks = ['A', 'B', 'C', 'D', 'E']
connections = [('A', 'B'), ('B', 'C'), ('C', 'D'), ('D', 'E'), ('E', 'A')]

# Add landmarks and connections to the graph
G.add_nodes_from(landmarks)
G.add_edges_from(connections)

# Draw the graph
pos = nx.circular_layout(G)
nx.draw(G, pos, with_labels=True, node_color='lightblue', node_size=2000, font_size=12, font_weight='bold')
nx.draw_networkx_edges(G, pos, edge_color='gray', width=2)

# Add title
plt.title("Hamiltonian Circuit in a City")

# Display the plot
plt.show()
```

In this example, the Hamiltonian circuit is represented by the cycle A -> B -> C -> D -> E -> A. This circuit visits each landmark (vertex) exactly once, forming a closed loop where the starting and ending point is the same. The Hamiltonian circuit satisfies all the properties mentioned, including containing every vertex, being connected, having the same number of edges as vertices, and every vertex having a degree of 2.
